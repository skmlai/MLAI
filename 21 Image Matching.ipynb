```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image1=cv2.imread('images/eiphel1.png',0)
image2=cv2.imread('images/eiphel2.png',0)
```

```python
plt.figure(figsize=(16,19))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(image1,cmap='gray')
plt.subplot(1,2,2)
plt.title('Thress Hold Image')
plt.imshow(image2,cmap='gray')
```

```python
sift=cv2.SIFT_create()
```

```python
keypoint_1, description_1=sift.detectAndCompute(image1,None)
keypoint_2, description_2=sift.detectAndCompute(image2,None)

len(keypoint_1), len(keypoint_2)
```

```python
# feature matching
bf=cv2.BFMatcher(cv2.NORM_L1)

matches = bf.match(description_1,description_2)
matches= sorted(matches,key=lambda x:x.distance)
img3=cv2.drawMatches(image1,keypoint_1,image2,keypoint_2,matches[:40],image2,flags=2)
plt.imshow(img3)
```
