```python
x=np.array([500,800,300,600,900,700,400])
```

```python
# define a function wich will return the scalled value
def scaler(x):
    x_new = ((x-x.min()) / (x.max() - x.min()))
    return x_new
```

```python
# Standardize
```

```python
x=np.array([500,800,300,600,900,700,400])
```

```python
def standardize(x):
    x_new=(x-x.mean())/x.std()
    return x_new
```

```python
x_new=standardize(x)
```

```python
x_new
```

```python
x_new.std()
```

```python
x_new.mean()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_scaled= scaler.fit_transform(x)
```

```python
x_scaled=pd.DataFrame(x_scaled, columns=x.columns)
x_scaled
```
