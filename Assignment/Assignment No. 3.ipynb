```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("images/sachin.jpg")
img1=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
img1.shape
```

```python
plt.imshow(img1)
```

```python
# Image Translation
height, width = image.shape[:2]
T=np.float32([[1,0,400],[0,1,-150]])
img_translation = cv2.warpAffine(image, T, (width,height))
```

```python
cv2.imshow("Original image", image)
cv2.imshow("Translated Image",img_translation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Image Rotation

center=(100,100) 
angle=-60 
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
rotated_image = cv2.warpAffine(image, rotation_matrix,(image.shape[1],image.shape[0]))
cv2.imshow("Rotated Image",rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resized_image=cv2.resize(img1,None, fx=0.75, fy=0.75) 
```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img1)
plt.subplot(1,2,2)
plt.title('Resized')
plt.imshow(resized_image)
```

```python
#Flip the image horizontally
flipped_horizontal = cv2.flip(img1,0)

#Flip the image vertically
flipped_vertical = cv2.flip(img1,1)

# Flip the image diagonally
flipped_diagonal = cv2.flip(img1,-1)

#Show the flipped images
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('Original Image')
plt.imshow(img1)
plt.subplot(1,4,2)
plt.title('Horizontally Flipped Image')
plt.imshow(flipped_horizontal)
plt.subplot(1,4,3)
plt.title('Vertically Flipped Image')
plt.imshow(flipped_vertical)
plt.subplot(1,4,4)
plt.title('Diagonally Flipped Image')
plt.imshow(flipped_diagonal)
```
