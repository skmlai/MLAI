```python
import cv2
# Load the Image
image = cv2.imread('images/lenna1.png')

# Flipp The Imaze horizental
flipped_horizental = cv2.flip(image,0)

# Flipp The Imaze verticaly
flipped_verticaly = cv2.flip(image,1)

# Flipp The Imaze Digonal
flipped_digonal = cv2.flip(image,-1)

# show the Image
cv2.imshow("Original Image",image)
cv2.imshow("Horizental Image",flipped_horizental)
cv2.imshow("Veritcal Image",flipped_verticaly)
cv2.imshow("Diagonal Image",flipped_digonal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flip_har = cv2.cvtColor(flipped_horizental, cv2.COLOR_BGR2RGB)
flip_Var = cv2.cvtColor(flipped_verticaly, cv2.COLOR_BGR2RGB)
flip_dia = cv2.cvtColor(flipped_digonal, cv2.COLOR_BGR2RGB)
img_org = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,17))
plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img_org)
plt.subplot(2,2,2)
plt.title("Horizental Image")
plt.imshow(flip_har)
plt.subplot(2,2,3)
plt.title("Vertical Image")
plt.imshow(flip_Var)
plt.title("Diagonal Image")
plt.subplot(2,2,4)
plt.imshow(flip_dia)
```
