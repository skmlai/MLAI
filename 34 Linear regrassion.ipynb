```python
## scale the data before the cross validation
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
x_scaled=scalar.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
# Cross Validation
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,random_state=0)
```

```python
x_train.shape,x_test.shape,x_train.shape,y_test.shape
```

```python
### Build the model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
###### Predict on the unseen data
```

```python
y_pred= classifier.predict(x_test)
```

```python
y_pred
```

```python
(y_pred-y_test).value_counts()
```

```python
### Performance Evaluation
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
```

```python
confusion_matrix(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm.diagonal().sum()/cm.sum()
```

```python
# Classification Report
```

```python
print(classification_report(y_test,y_pred))
```

```python
# recall-0 / accuracy of class 0 / specificity of algorithm
r0=67/68
r0
```

```python
# recall-1/ accuracy of class 0 / specificity of algorithm
r1=22/32
r1
```

```python
# precision-0 / Positive Predictive Value
p0=67/77
p0
```

```python
# precision-1 / Negative Predictive Value
p1=22/23
p1
```

```python
# F1 - score - 0
2*p0 * r0 / (p0+r0)
```

```python
# F1 - score - 1
2*p1 * r1 / (p1+r1)
```

```python
classifier.predict_proba(x_test)
```
