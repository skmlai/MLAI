# Rotation Using Matrix

```python
import cv2

# Load the Image
image = cv2.imread('images/lenna.png')

# Define the Centre and Rotation angle
centre = (image.shape[1]//2, image.shape[0]//2)
angle=60

# Rotate Image
rotated_matrix = cv2.getRotationMatrix2D(centre,angle,1.0)

#apply the rotation using warpAffine
img_translation = cv2.warpAffine(image, rotated_matrix, (image.shape[1], image.shape[0]))
# Show the rotated Image
cv2.imshow("Rotate Image",img_translation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
