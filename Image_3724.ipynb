```python
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('images/lenna.png',0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,4,1)
plt.title('Gray')
plt.imshow(img, cmap='gray')
plt.subplot(2,4,2)
plt.title('Blues')
plt.imshow(img, cmap='Blues')
plt.subplot(2,4,3)
plt.title('Greens')
plt.imshow(img, cmap='Greens')
plt.subplot(2,4,4)
plt.title('Reds')
plt.imshow(img, cmap='Reds')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,4,1)
plt.title('bwr')
plt.imshow(img, cmap='bwr')
plt.subplot(2,4,2)
plt.title('BrBG')
plt.imshow(img, cmap='BrBG')
plt.subplot(2,4,3)
plt.title('RdGy')
plt.imshow(img, cmap='RdGy')
plt.subplot(2,4,4)
plt.title('RdBu')
plt.imshow(img, cmap='RdBu')
plt.subplot(2,4,5)
plt.title('YlOrRd')
plt.imshow(img, cmap='YlOrRd')
```
