```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the Image
image = cv2.imread('images/lenna1.jpg')

# Resize The Imaze to a fixed width and size
resize_image = cv2.resize(image,(700,500))

# show the resize Image
cv2.imshow("Rotate Image",resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resizing Using Scalling Factor
```

```python
# Load the Image
image = cv2.imread('images/lenna1.jpg')

# Resize The Imaze to a fixed width and size
resize_image = cv2.resize(image,None,fx=1.75,fy=1.75)

# show the resize Image
cv2.imshow("Rotate Image",resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from skimage.transform import resize
import matplotlib.pyplot as plt
# Load the Image
image = cv2.imread('images/lenna1.png')

# Resize The Imaze with billinear interpolation
resize_image = resize(image,(1000,1200),anti_aliasing=True)

# show the resize Image
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resize_image)
```
